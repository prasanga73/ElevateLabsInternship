```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-5, 6)
x = np.where((n >= 0) & (n < 4), 2, 0)
x_neg = np.where((-n >= 0) & (-n < 4), 2, 0)

x_even = (x + x_neg) / 2
x_odd = (x - x_neg) / 2

# plt.stem(n, x, basefmt="C0-", linefmt='C0-', markerfmt='C0o', label='x[n]')
# plt.stem(n, x_even, basefmt="C1-", linefmt='C1-', markerfmt='C1x', label='Even component $x_e[n]$')
plt.stem(n, x_odd, basefmt="C2-", linefmt='C2-', markerfmt='C2s', label='Odd component $x_o[n]$')

plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Signal and its Even and Odd Components')
plt.legend()
plt.grid(True)
plt.show()
```
