```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.describe()
```

```python
numeric_class = df.select_dtypes(include='number').columns.tolist()
print(numeric_class)
```

```python
for classes in numeric_class:
    sns.histplot(df[classes],bins=30,kde=True)
    plt.title(f'Histogram of {classes}')
    plt.xlabel(classes)
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()
    
    
```

```python
for classes in numeric_class:
    sns.boxplot(df[classes])
    plt.title(f'Box plot of {classes}')
    plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```

```python
  
sns.pairplot(df[numeric_class],hue='Survived',diag_kind='kde')
```
